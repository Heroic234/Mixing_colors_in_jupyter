# Mixing colors



Importing the needed libraries

```python
import numpy as np
from matplotlib import pyplot as plt
```

We can mix primary colors red(r),green(g), and blue(b) with the specification color=(r,g,b) where r,g,b are values of red,green, and blue in the mix with values ranging from 0 to 1.
For example color = (1,0,0) gives pure red; color=(1,0,1) gives magenta, a purplish mix of red and blue; color=(0,0,0) gives black and color(1,1,1) gives white

```python
plt.axis([0,100,0,20]) # parameters for the x-axis axis and y-axis

for x in np.arange(1,100,1): # creating a loop with numbers between 1 & 100 in intervals of 1
        r=x/100  
        g=0
        b=0
        plt.plot([x,x],[0,20],linewidth=5,color=(r,g,b))
        
plt.show()
 
```

By altering the parameters we can change intensity of colors and even mix them

```python
#Mixing red and green
plt.axis([0,100,0,20])

for x in np.arange(1,100,1):
    r=x/100
    g=x/100
    b=0
    plt.plot([x,x],[0,20],linewidth=5,color=(r,g,b))
plt.show()
```

```python
#Mixing red and blue
plt.axis([0,100,0,20])

for x in np.arange(1,100,1):
    r=x/100
    g=0
    b=x/100
    plt.plot([x,x],[0,20],linewidth=5,color=(r,g,b))
plt.show()
```

```python
#Mixing red and blue with a hint of mathematics
plt.axis([0,100,0,20]) # parameters for the x-axis axis and y-axis

for x in np.arange(1,100,1):                  # creating a loop with numbers between 1 & 100 in intervals of 1
        r=x/100
        g=0
        b=np.ceil(x/100)                      #using the ceil function we are able to obtain the smallest value from all loop instances of x/100 
        plt.plot([x,x],[0,20],linewidth=5,color=(r,g,b))   # using this idealogy we can create a continuous mixture that increase/decreases per x inputs using trig 
        
plt.show()
```

```python
plt.axis([0,100,0,20])

for x in np.arange(1,100,1):
    r=x/1000                               #a mixture of red,green and blue
    g=x/100
    b=x/1000
    plt.plot([x,x],[0,20],linewidth=5,color=(r,g,b))
plt.show()
```

```python
plt.axis([0,100,0,20]) #Mixing all there colors in low quantities

for x in np.arange(1,100,1):
    r=np.cos(np.floor(x/100)) #flooring the value allows us to use only the highest values
    g=np.ceil(x/100)
    b=np.sin(x/100)
    plt.plot([x,x],[0,20],linewidth=5,color=(r,g,b))
plt.show()
```

```python
plt.axis([0,100,0,20])

for x in np.arange(1,100,1):
    r=abs(np.sin(x))
    g=x/100
    b=1
    plt.plot([np.tanh(x),4*x],[0,20],linewidth=5,color=(r,g,b))
plt.show()
```

```python
plt.axis([0,100,0,20])

for x in np.arange(1,100,1):
    r=abs(np.cos(x)*np.sin(x))
    g=x/100
    b=1
    plt.plot([4*x,np.tanh(x)],[0,20],linewidth=5,color=(r,g,b))
plt.show()
```

```python
plt.axis([0,100,0,20])

for x in np.arange(1,100,1):
    r=abs(np.sin(x**2/100))*np.tanh(x) #taking absolute values removes the negative values since color intensity is controlled with positve values
    y=x/100
    b=1
    plt.plot([x,np.tanh(x)],[0,20],linewidth=5,color=(r,y,b))
plt.show()
# this was an experimental effect that surprised me....
#but from my analysis keeping one color value constant and varing the inputs of the other with trig functions
#over discrete intervals has the capability of producing discrete shades that dissappear and reappear based on the output of these functions
```

# controlling color intensity


We can use the alpha attriute to control color intensity; alpha has values between 0-1. .scatter() helps us to plot dots and plt.arrow() for arrows

```python

plt.scatter(80,50,s=10000,color='b',alpha=1) #s = size of dots  #b = color code for blue
plt.scatter(90,50,s=10000,color='b',alpha=0.5)
plt.scatter(100,50,s=10000,color='b',alpha=0.1,)



plt.show()
```
